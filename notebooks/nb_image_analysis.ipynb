```python
import os
import gzip
import shutil
from PIL import Image
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_dir = 'data/train/'
label_dir = 'data/label/'
train_files = [train_dir + file for file in os.listdir(train_dir)]
label_files = [label_dir + file for file in os.listdir(label_dir)]

print(len(train_files))
print(len(label_files))
```

```python
suffix = 'extract'

# extraction ranges
x_idx_range = slice(0, 32)
y_idx_range = slice(0, 32)
z_idx_range = slice(0, 32)
```

# Train slices

```python
for f in train_files:
    name = f
    new_name = name.replace('data/train/','')
    #print(name)
    #print(new_name)
    nii_file  = nib.load(f)
    nii_data = nii_file.get_fdata()
    #print(type(nii_data))
    print(nii_data.shape)
    
    block = np.asarray(nii_file.dataobj[x_idx_range, y_idx_range, z_idx_range])
    #print(type(block))
    print(block.shape)
    
    filename = 'data_processed/x_images/' + new_name.replace('campus','')
    print(filename)

    img = nib.Nifti1Image(block, affine=nii_file.affine)
    basename, ext = nii_file.get_filename().split(os.extsep, 1)
    out_name = '{}_{}.{}'.format(basename, suffix, ext)
    nib.save(img, filename)
    print('')
```

# Label slices

```python
for f in label_files:
    name = f
    new_name = name.replace('data/label/','')
    #print(name)
    #print(new_name)
    nii_file  = nib.load(f)
    nii_data = nii_file.get_fdata()
    #print(type(nii_data))
    print(nii_data.shape)
    
    block = np.asarray(nii_file.dataobj[x_idx_range, y_idx_range, z_idx_range])
    #print(type(block))
    print(block.shape)
    
    filename = 'data_processed/y_images/' + new_name.replace('campus','')
    #print(filename)

    img = nib.Nifti1Image(block, affine=nii_file.affine)
    basename, ext = nii_file.get_filename().split(os.extsep, 1)
    out_name = '{}_{}.{}'.format(basename, suffix, ext)
    nib.save(img, filename)
    print('')
```
